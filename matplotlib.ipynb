```python
import pandas as pd
import matplotlib.pyplot as pt
import numpy as np
```

```python
f=pd.read_csv("Mall_Customers.csv")
```

```python
f.head()
```

## Line graph

```python
pt.figure(figsize=(20,7),facecolor="g",)
pt.plot(f.Age,marker="*",markersize=15,markerfacecolor="r")
pt.title("Index vs Age",fontsize=25,c="b")
pt.xticks([x for x in range(0,210,5)],fontsize=12,c="b")
pt.yticks([x for x in range(0,85,5)],fontsize=12,c="b")
pt.xlabel("Index",fontsize=20,c="b")
pt.ylabel("Age",fontsize=20,c="b")
pt.savefig("indexvsage.png")
pt.grid()
pt.show()
```

## Histogram

```python
pt.figure(figsize=(20,7),facecolor="g")
pt.hist(f.Age,bins=13,color="b",histtype="barstacked")
pt.xticks([x for x in range(18,75,4)],fontsize=15,c="b")
pt.yticks(fontsize=15,c="b")
pt.xlabel("Age group",fontsize=20,c="b")
pt.ylabel("no. of people",fontsize=20,c="b")
pt.title("People vs age-group",fontsize=25,c="r")
pt.grid()
pt.show()
```

## Bar graph

```python
f=pd.read_csv("summer_2012.csv")
```

```python
f.head()
```

```python
pt.figure(figsize=(20,7),facecolor="g")
pt.bar(f.Country,f.Medal,width=0.7)
pt.xticks(fontsize=19,c="white")
pt.yticks(fontsize=14,c="white")
pt.xlabel("Country Name",fontsize=20,c="b")
pt.ylabel("no. Medal",fontsize=20,c="b")
pt.title("Country vs Medal",fontsize=25,c="r")
pt.grid()
pt.show()
```

```python
pt.figure(figsize=(10,20),facecolor="g")
pt.barh(f.Country,f.Medal)
pt.xticks(fontsize=19,c="white")
pt.yticks(fontsize=14,c="white")
pt.ylabel("Country Name",fontsize=20,c="b")
pt.xlabel("no. Medal",fontsize=20,c="b")
pt.title("Country vs Medal",fontsize=25,c="r")
pt.grid()
pt.show()
```

## Scatter graph

```python
f=pd.read_csv("Mall_Customers.csv")
```

```python
f.head()
```

```python
pt.figure(figsize=(20,7),facecolor="g")
pt.scatter(f.Age,f["Annual Income (k$)"],c="b",marker="x",alpha=0.9,)
pt.ylabel("Anual income",fontsize=20,c="b")
pt.xlabel("Age",fontsize=20,c="b")
pt.title("Age vs Anual-income",fontsize=25,c="r")
pt.xticks(fontsize=19,c="white")
pt.yticks(fontsize=19,c="white")
pt.grid()
pt.show()
```

## Pie graph

```python
d={"Name":["Shrish","Kalyani","Arvind","Harshvardhan"],
  "Marks":[67,76,55,68]}
f=pd.DataFrame(d)
f
```

```python
pt.pie(f.Marks,shadow=True,colors=["red","blue","green","brown"],labels=f.Name,autopct='%1.2f%%',radius=2,labeldistance=0.5,explode=[0.1,0.1,0.1,0.1])
pt.show()
```
